```python
menu = {
    "sushi": 10.99,
    "pepsi": 2.99,
    "pizza" : 9.99,
}

tax_rate = 0.07

def display_menu(menu):
    print("-----------Restaurant Menu------------")
    print()
    for item,price in menu.items():
        print(f"{item.title():<13}   ${price}")

    print()
    print("-"*40)

def calculate_totals(menu,order):
    subtotal = 0
    for item,quantity in order.items():
        price = menu[item]
        subtotal += price * quantity
    tax = subtotal * tax_rate 
    total = subtotal + tax 
    return subtotal,tax,total

def print_receipt(customer_name,menu,order,subtotal,tax,total):
    print("===============Receipt===============")
    print(f"Customer : {customer_name}")

    print()
    for item,quantity in order.items():
        price = menu[item]
        line_total = price * quantity 
        print(f"{item.title():<15}{quantity:<5}{price:<8}{line_total:.2f}")

    print("-"*40)
    print(f"Subtotal:         ${subtotal:.2f}")
    print(f"Tax:              ${tax:.2f}")
    print(f"Total:            ${total:.2f}")

    print(f"Thank you for your order,Have a nice day")
    print("="* 40)
def main():
    print("Welcome to UKP burmese restaurent")
    customer_name=input("Please enter your name: ").strip()

    order = {}
    display_menu(menu)

    while True:
        item = input("Type an item from the menu or 'done'to finish the order?: ").strip()

        if item == "done":
            break
        if item not in menu:
            print("Sorry,that item is not on the menu,please try again.")
            continue

        quantity_input= input("How many would you like to order?: ")
        if not quantity_input.isdigit():
            print("Quantity must be a positive whole number.Try again.")
            continue

        quantity = int(quantity_input)
        if quantity <= 0:
            print("Quantity must be at least 1.Try again.")
            continue
        if item in order:
            order[item] += quantity 
        else:
            order[item] = quantity 

        print(f"Added {quantity} {item} to your order.")

    if not order:
        print("You did not order anything.Goodbye.")
        return


    subtotal,tax,total = calculate_totals(menu,order)
    print_receipt(customer_name,menu,order,subtotal,tax,total)

main()

        


 
```
